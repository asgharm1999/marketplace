```python
import pandas as pd
import numpy as np
import datetime
from bs4 import BeautifulSoup
import urllib3
import requests
import time
import re
import os
import math
from urllib.request import urlopen
```

```python
#Scrape AptDeco-----------------------------------------------------------------------------------------------------------

base_url_aptdeco = "https://www.aptdeco.com/catalog/furniture?region=Northeast+%28NY%2C+NJ%2C+CT%2C+PA%2C+DE%29"

html_aptdeco = urlopen(base_url_aptdeco)
bsyc_aptdeco = BeautifulSoup(html_aptdeco.read(), "lxml")
fout_aptdeco = open('aptdeco_html.csv', 'wt', encoding='utf-8')
fout_aptdeco.write(str(bsyc))
fout_aptdeco.close()

page_aptdeco = requests.get(base_url_aptdeco)
#print(page.content)
```

```python
soup_aptdeco = BeautifulSoup(page_aptdeco.content, "html.parser")

num_results_aptdeco = str(soup_aptdeco.find('div', 'Stats__StyledDiv-vrpu8w-0 fuknnX').text)
num_results_aptdeco = num_results_aptdeco[:-8]
num_results_aptdeco = int(num_results_aptdeco)
print(num_results_aptdeco)

product_results_aptdeco = soup_aptdeco.find_all('div', {'class': 'styles__ProductGridCell-sc-1859r2o-0 bFDJTu'})

for product in product_results_aptdeco:
    print(product.prettify())
```

```python
products_info_aptdeco = []

for product in product_results_aptdeco:
    title = product.find('div', {'class': 'Card__ItemName-rr6223-2 hKSDjR'}).text
    product_url = 'https://www.aptdeco.com' + product.find('a', {'class': 'Card__CardLink-rr6223-1 gfTIJo'})['href']
    new_price = product.find('div', {'class': 'Flex__Container-sc-14qf74e-0 Flex__Row-sc-14qf74e-1 dVjRzf djUejJ'}).text
    separator_new_index = new_price.index('â€¢')-1
    if product.find('s', {'class': 'Card__StrikeThrough-rr6223-3 iwsrff'}) != None:
        new_price_index = new_price.index('$',1)
        new_price = new_price[new_price_index:separator_new_index]
    else:
        separator_new_index = new_price.index('â€¢')-1
        new_price = new_price[:separator_new_index]
#    if product.find('li', {'class': 'Card__OriginalPrice-rr6223-5 kXuMgE'}) != None:
#        retail_price = product.find('li', {'class': 'Card__OriginalPrice-rr6223-5 kXuMgE'}).text
#        separator_retail_index = retail_price.index('|') 
#        retail_price = retail_price[12:separator_retail_index]
#    else:
#        retail_price = None
    
    product_page = requests.get(product_url)
    product_soup = BeautifulSoup(product_page.content, "html.parser")
    location = product_soup.find('p', {'class': 'PDPFooter__FooterText-sc-1ngdbbr-5 PDPFooter__SpacedFooterText-sc-1ngdbbr-6 ddMTKT CcsNI'}).text
    separator_location_index = location.index(':')+2
    location = location[separator_location_index:]
    
    products_info_aptdeco.append([product_url, new_price, location, title])

columns = ('Post URL', 'Price', 'Location', 'Post Title')
df = pd.DataFrame(products_info_aptdeco, columns=columns)

timestamp = datetime.datetime.now().strftime('%m_%d_%y %H%M%S')
df.to_csv(f'AptDeco Results ({timestamp}).csv', index=False)
    
for info in products_info_aptdeco:
    print(info)
```

```python
#Scrape Uhaul-----------------------------------------------------------------------------------------------------------

base_url_uhaul = "https://www.uhaul.com/Locations/Pittsburgh-PA/Results/"

html_uhaul = urlopen(base_url_uhaul)
bsyc_uhaul = BeautifulSoup(html_uhaul.read(), "lxml")
fout_uhaul = open('uhaul_html.txt', 'wt', encoding='utf-8')
fout_uhaul.write(str(bsyc_uhaul))
fout_uhaul.close()

page_uhaul = requests.get(base_url_uhaul)
#print(page.content)
```

```python
soup_uhaul = BeautifulSoup(page_uhaul.content, "html.parser")

store_results_uhaul = soup_uhaul.find_all('li', {'class': 'divider'})

for store in store_results_uhaul:
    print(store.prettify())
```

```python
store_info_uhaul = []

for store in store_results_uhaul:
    name = store.find('h3', {'class': 'collapse-half medium-uncollapse'}).text
    if(store.find('small', {'class': 'text-semibold'})) != None:
        name = name[3:-55]
    else:
        name = name[3:-3]
    url = store.h3.a['href']
    location = store.find('p', {'class': 'collapse'}).text
    location = location[23:-23]
    contact = store.find('ul', {'class': 'no-bullet collapse'}).find('a').text
    contact = contact[30:-26]
    open_hours = store.find('p', {'class': 'text-callout text-semibold text-lg'}).text
    open_hours = open_hours[10:-6]
    
    store_info_uhaul.append([url, name, location, contact, open_hours])

columns = ('Store URL', 'Name', 'Location', 'Contact No', 'Operating Hours')
df = pd.DataFrame(store_info_uhaul, columns=columns)

timestamp = datetime.datetime.now().strftime('%m_%d_%y %H%M%S')
df.to_csv(f'Uhaul Results ({timestamp}).csv', index=False)    

for info in store_info_uhaul:
    print(info)
```
